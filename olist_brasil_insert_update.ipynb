### O que é ETL?


A ETL (Extração, Transformação e Carga) é um processo utilizado para integrar dados de diferentes fontes, transformá-los em um formato que possa ser analisado e carregá-los em um novo destino, geralmente um data warehouse ou um data lake. Seguem abaixo os passos para realizar uma ETL:

* Identificar as fontes de dados: 

Identifique todas as fontes de dados que serão usadas na ETL, como bancos de dados, planilhas, arquivos CSV, feeds RSS, etc.

* Extrair os dados: 

Extrair os dados das fontes identificadas. Isso pode ser feito usando uma variedade de ferramentas, dependendo da fonte de dados. Por exemplo, bancos de dados podem ser extraídos usando SQL, enquanto planilhas podem ser extraídas usando bibliotecas Python como Pandas.

* Validar e limpar os dados: 

Após extrair os dados, é importante validá-los e limpá-los para garantir que os dados estejam completos, precisos e consistentes. Isso envolve remover dados duplicados ou inválidos e preencher quaisquer valores ausentes ou incorretos.

* Transformar os dados: 

Transforme os dados em um formato que possa ser carregado no destino. Isso pode envolver a conversão de tipos de dados, a criação de novas colunas ou a agregação de dados para criar novas visões.

* Analisar dados: 

Uma vez que os dados são limpos e transformados, é importante realizar uma análise exploratória para entender melhor os dados. Isso pode incluir a criação de gráficos, a identificação de padrões e a aplicação de outras técnicas de análise.

* Selecionar recursos: 

Em muitos casos, nem todos os dados são relevantes para o problema que se está tentando resolver. É importante selecionar os recursos mais importantes para o modelo que se está tentando construir.

* Dividir os dados: 

Antes de construir o modelo final, é importante dividir os dados em um conjunto de treinamento e um conjunto de teste. O conjunto de treinamento é usado para construir o modelo, enquanto o conjunto de teste é usado para avaliar a precisão do modelo.

* Verificar qualidade: 

Após dividir os dados, é importante verificar a qualidade dos dados para garantir que eles estejam prontos para serem usados no modelo final. Isso pode incluir a identificação de valores ausentes ou inconsistentes, a verificação de distribuições de variáveis e outras técnicas de validação de dados.

* Preparar dados para o modelo: 

Finalmente, é importante preparar os dados para serem usados no modelo final. Isso pode incluir a codificação de variáveis categóricas, a normalização de dados e outras técnicas de preparação de dados específicas para o tipo de modelo que você está construindo.

* Carregar os dados: 

Finalmente, carregue os dados transformados no destino, como um data warehouse ou um data lake. Isso pode ser feito usando ferramentas ETL dedicadas, como o Apache Airflow, Pentaho ou o Apache Nifi.

* Monitorar e manter: 

Após carregar os dados, monitore e mantenha o processo ETL para garantir que os dados sejam atualizados regularmente e que quaisquer problemas sejam resolvidos imediatamente. Isso pode envolver o uso de ferramentas de monitoramento e alerta para identificar problemas rapidamente e tomar medidas corretivas.

### Arquitetura do sistema: 

* Descreva a arquitetura geral do sistema de engenharia de dados. Inclua uma visão geral dos componentes principais, como pipelines de ingestão de dados, armazenamento de dados, processamento e transformação de dados, bem como os sistemas de consulta e análise.

A arquitetura geral do sistema de engenharia de dados mencionado envolve os seguintes componentes principais:

* Pipelines de ingestão de dados: A partir do site da Kaggle, será realizada a coleta dos dados disponibilizados. Esses dados serão inicialmente armazenados em uma camada transitória (transient) para fins de processamento e verificação inicial da qualidade.

> O uso de arquivos no formato Parquet traz benefícios significativos em um sistema de engenharia de dados, incluindo eficiência de armazenamento, com compressão e compactação eficientes, e performance aprimorada para leitura e consulta, permitindo a leitura seletiva de colunas e suporte à leitura em paralelo. Além disso, o Parquet possui um esquema de dados incorporado, facilitando a interpretação dos dados, e é amplamente suportado por frameworks de processamento, garantindo interoperabilidade e integração com outras ferramentas e ecossistemas de dados. No geral, o uso de arquivos Parquet proporciona um sistema de engenharia de dados mais eficiente, escalável e compatível com diferentes necessidades analíticas.

* Processamento e transformação de dados: Na fase de refinamento (refined), os dados passarão por processamentos e transformações adicionais, como limpeza, normalização, agregação, enriquecimento e outras etapas de preparação para análise. Durante esse processo, os dados refinados serão armazenados em um banco de dados Postgres, que será utilizado para consultas e análises posteriores.

* Consulta e análise: O banco de dados Postgres servirá como fonte de dados para sistemas de consulta e análise. A partir desse banco de dados, os usuários ou outros sistemas poderão realizar consultas complexas, executar análises estatísticas, criar visualizações e extrair insights relevantes para o negócio.

* Essa arquitetura em camadas permite a organização e o processamento gradual dos dados, desde a ingestão inicial até a disponibilização para consulta e análise. O uso do formato Parquet e o armazenamento em um banco de dados Postgres contribuem para melhorar a performance e a eficiência do sistema, garantindo que os dados estejam prontos e acessíveis para as análises e tomadas de decisões necessárias.

> Convertendo os arquivos csv carregados na transient em arquivos parquet e carregando na , após carregar a raw, será eliminado os arquivos da transient.

* OBS: Esse código foi elaborado para realizar a primeira carga na raw caso não exista arquivos neste. Em segundo momento ele estará concatenando os novos dados e truncando as duplicadas com base nas PK's... Simulanto o inserte/update dos dados.

```python
def carregando_raw():
    folder = os.getcwd()
    file_path = os.path.join(folder, 'data/controller.csv')
    path = pd.read_csv(file_path)
    print('Path criado com base no arquivo controller!')

    for i in range(len(path['path_transient'])):
        # Criando o caminho dos arquivos e diretórios
        diretorio_transient = path['path_transient'][i]
        nome_arquivo_transient = path['table_transient'][i] + '.csv'
        caminho_arquivo_transient = (folder + diretorio_transient + nome_arquivo_transient)

        diretorio_raw = path['path_raw'][i]
        nome_arquivo_raw = path['table_raw'][i] + '.parquet'
        caminho_arquivo_raw = (folder + diretorio_raw + nome_arquivo_raw)

        # Verificando a existência dos arquivos em Transient e carrega na Raw pela primeira vez
        if os.path.exists(caminho_arquivo_transient) and not os.path.exists(caminho_arquivo_raw):
            # Carregando arquivos csv
            df = pd.read_csv(caminho_arquivo_transient)

            # Dropando os duplicados preservando o último atualizado
            df.drop_duplicates(subset=df.columns[0], keep='last', inplace=True)

            # Verifica se o DataFrame não está vazio
            if not df.empty:
                # Salva na Raw convertendo para o formato Parquet
                df.to_parquet(caminho_arquivo_raw)
                print(f"Arquivo Parquet {nome_arquivo_raw} criado: {caminho_arquivo_raw}")
            else:
                print(f"O DataFrame está vazio: {nome_arquivo_raw}")

        # Caso a Raw já tenha sido carregada pela primeira vez, esse código é executado para atualizar os arquivos verificando se algum arquivo foi deletado.
        elif os.path.exists(caminho_arquivo_transient) and os.path.exists(caminho_arquivo_raw):
            # Confere a existência do arquivo, caso deletado, cria novamente
            if not os.path.exists(caminho_arquivo_raw):
                df = pd.read_csv(caminho_arquivo_transient)
                df.drop_duplicates(subset=df.columns[0], keep='last', inplace=True)
                if not df.empty:
                    df.to_parquet(caminho_arquivo_raw)
                    print(f"Arquivo Parquet {nome_arquivo_raw} criado: {caminho_arquivo_raw}")

            # Carregando os DataFrames para concatenar os novos dados
            df1 = pd.read_csv(caminho_arquivo_transient)
            df2 = pd.read_parquet(caminho_arquivo_raw)

            # Concatenando os dados
            df_concatenado = pd.concat([df2, df1], axis=0)

            # Eliminando as duplicatas pelo ID mantendo sempre a última ocorrência
            df_concatenado = df_concatenado.drop_duplicates(subset=df_concatenado.columns[0], keep='last')
            df_concatenado.to_parquet(caminho_arquivo_raw)
        else:
            print(f'Arquivo Raw {nome_arquivo_raw} já existe! Para atualizá-lo, carregue a transient correspondente.')

    # Após a criação e atualização dos dados, os arquivos Transient são eliminados deixando o diretório
    # livre para receber novos dados
    print()
    print()
    for i in range(len(path['path_transient'])):
        diretorio_transient = path['path_transient'][i]
        nome_arquivo_transient = path['table_transient'][i] + '.csv'
        caminho_arquivo_transient = (folder + diretorio_transient + nome_arquivo_transient)
        
        if os.path.exists(caminho_arquivo_transient):
            os.remove(caminho_arquivo_transient)

        print(f"Arquivo {file_path} da Transient eliminado.")

carregando_raw()
```

```python
def carregando_trusted():
    folder = os.getcwd()
    file_path = os.path.join(folder, 'data/controller.csv')
    path = pd.read_csv(file_path)

    for i in range(len(path['path_raw'])):
        # Criando o caminho dos arquivos e diretórios
        diretorio_raw = path['path_raw'][i]
        nome_arquivo_raw = path['table_raw'][i] + '.parquet'
        caminho_arquivo_raw = (folder + diretorio_raw + nome_arquivo_raw)

        diretorio_trusted = path['path_trusted'][i]
        nome_arquivo_trusted = path['table_trusted'][i] + '.parquet'
        caminho_arquivo_trusted = (folder + diretorio_trusted + nome_arquivo_trusted)

        # Verificando a existência dos arquivos em Transient e carrega na Raw pela primeira vez
        if os.path.exists(caminho_arquivo_raw) and not os.path.exists(caminho_arquivo_trusted):
            # Carregando arquivos parquet
            df = pd.read_parquet(caminho_arquivo_raw)

            # Dropando os duplicados preservando o último atualizado
            df.drop_duplicates(subset=df.columns[0], keep='last', inplace=True)

            # Verifica se o DataFrame não está vazio
            if not df.empty:
                # Salva na Raw convertendo para o formato Parquet
                df.to_parquet(caminho_arquivo_trusted)
                print(f"Arquivo Parquet {nome_arquivo_trusted} criado: {caminho_arquivo_trusted}")
            else:
                print(f"O DataFrame está vazio: {nome_arquivo_trusted}")

        # Caso a Raw já tenha sido carregada pela primeira vez, esse código é executado para atualizar os arquivos verificando se algum arquivo foi deletado.
        elif os.path.exists(caminho_arquivo_raw) and os.path.exists(caminho_arquivo_trusted):
            # Confere a existência do arquivo, caso deletado, cria novamente
            if not os.path.exists(caminho_arquivo_trusted):
                df = pd.read_parquet(caminho_arquivo_raw)
                df.drop_duplicates(subset=df.columns[0], keep='last', inplace=True)
                if not df.empty:
                    df.to_parquet(caminho_arquivo_trusted)
                    print(f"Arquivo Parquet {nome_arquivo_trusted} criado: {caminho_arquivo_trusted}")

            # Carregando os DataFrames para concatenar os novos dados
            df1 = pd.read_parquet(caminho_arquivo_raw)
            df2 = pd.read_parquet(caminho_arquivo_trusted)

            # Concatenando os dados
            df_concatenado = pd.concat([df2, df1], axis=0)

            # Eliminando as duplicatas pelo ID mantendo sempre a última ocorrência
            df_concatenado = df_concatenado.drop_duplicates(subset=df_concatenado.columns[0], keep='last')
            df_concatenado.to_parquet(caminho_arquivo_trusted)
        else:
            print(f'Arquivo Raw {nome_arquivo_trusted} já existe! Para atualizá-lo, carregue a transient correspondente.')
carregando_trusted()
```

### Modelagem dos dados: 

No projeto em questão, foi adotada uma estrutura de dados baseada em um modelo de banco de dados relacional. O modelo utilizado é composto por várias entidades, atributos e relacionamentos que representam os diferentes elementos do sistema de comércio eletrônico.

As principais entidades do modelo são:

* dim_customers: Representa os clientes do comércio eletrônico. Possui atributos como pk_customer (chave primária), unique_id (identificador único) e pk_zip_code (chave estrangeira para a tabela dim_geolocation).

* dim_geolocation: Representa a geolocalização, contendo informações sobre latitude, longitude, cidade, estado e pk_zip_code (chave primária).

* dim_products: Representa os produtos disponíveis para venda. Possui atributos como pk_product (chave primária), category_name, photos_qty (quantidade de fotos), além de dimensões físicas como peso, comprimento, altura e largura.

* dim_sellers: Representa os vendedores. Contém atributos como pk_seller (chave primária) e pk_zip_code (chave estrangeira para dim_geolocation).

* fat_items: Representa os itens dos pedidos, com atributos como pk_order (chave primária composta com pk_item), pk_product (chave estrangeira para dim_products), pk_seller (chave estrangeira para dim_sellers), shipping_limit_date (data limite de envio), price (preço) e freight (frete).

* fat_orders: Representa os pedidos realizados pelos clientes. Possui atributos como pk_order (chave primária), pk_customer (chave estrangeira para dim_customers), status (status do pedido), e datas relevantes como purchase_timestamp (data de compra), approved_at (data de aprovação), delivered_carrier_date (data de entrega pelo transportador), delivered_customer_date (data de entrega ao cliente) e estimated_delivery_date (data estimada de entrega).

* fat_payments: Representa os pagamentos relacionados aos pedidos. Contém atributos como pk_order (chave primária, referenciando fat_orders), sequential (número sequencial do pagamento), type (tipo de pagamento), installments (número de parcelas) e value (valor do pagamento).

* fat_reviews: Representa as avaliações feitas pelos clientes sobre os pedidos. Possui atributos como pk_review (chave primária), pk_order (chave estrangeira para fat_orders), score (pontuação), comment_title (título do comentário), comment_message (mensagem do comentário), creation_date (data de criação) e answer_timestamp (data de resposta).
